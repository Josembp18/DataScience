```python
import pandas as pd

print("=== JOINS EN PYTHON: INNER, LEFT, RIGHT JOIN ===\n")

# 1. CREACIÓN DE DATAFRAMES DE EJEMPLO
print("1. DATAFRAMES DE EJEMPLO:")

# DataFrame de empleados
empleados = pd.DataFrame({
    'empleado_id': [1, 2, 3, 4, 5],
    'nombre': ['Ana', 'Carlos', 'María', 'Pedro', 'Laura'],
    'departamento_id': [101, 102, 101, 103, 104]
})

# DataFrame de departamentos
departamentos = pd.DataFrame({
    'departamento_id': [101, 102, 103, 105],
    'nombre_departamento': ['Ventas', 'IT', 'Marketing', 'Finanzas'],
    'presupuesto': [50000, 75000, 30000, 60000]
})

print("DataFrame EMPLEADOS:")
print(empleados)
print(f"Shape: {empleados.shape}\n")

print("DataFrame DEPARTAMENTOS:")
print(departamentos)
print(f"Shape: {departamentos.shape}")

print("\n" + "="*60)
```

```python
# 2. INNER JOIN (UNIÓN INTERNA)
print("\n2. INNER JOIN - Solo registros que coinciden en AMBAS tablas")

# Realizar el inner join
inner_join = pd.merge(empleados, departamentos, on='departamento_id', how='inner')
print("Resultado del INNER JOIN:")
print(inner_join)
print(f"Shape: {inner_join.shape}")

# Explicación detallada
print("\n💡 EXPLICACIÓN:")
print("• Solo muestra empleados que tienen un departamento válido")
print("• Excluye a Laura (departamento_id 104 no existe en departamentos)")
print("• Excluye Finanzas (departamento_id 105 no tiene empleados)")
print("• Resultado: 4 filas (de 5 empleados y 4 departamentos)")

print("\n" + "="*60)
```

```python
# 3. LEFT JOIN (UNIÓN IZQUIERDA)
print("\n3. LEFT JOIN - Todos los registros de la tabla IZQUIERDA (empleados)")

# Realizar el left join
left_join = pd.merge(empleados, departamentos, on='departamento_id', how='left')
print("Resultado del LEFT JOIN:")
print(left_join)
print(f"Shape: {left_join.shape}")

# Explicación detallada
print("\n💡 EXPLICACIÓN:")
print("• Mantiene TODOS los empleados (tabla izquierda)")
print("• Laura aparece aunque su departamento_id 104 no existe")
print("• Los valores faltantes se llenan con NaN")
print("• Resultado: 5 filas (todos los empleados)")

print("\n" + "="*60)
```

```python
# 4. RIGHT JOIN (UNIÓN DERECHA)
print("\n4. RIGHT JOIN - Todos los registros de la tabla DERECHA (departamentos)")

# Realizar el right join
right_join = pd.merge(empleados, departamentos, on='departamento_id', how='right')
print("Resultado del RIGHT JOIN:")
print(right_join)
print(f"Shape: {right_join.shape}")

# Explicación detallada
print("\n💡 EXPLICACIÓN:")
print("• Mantiene TODOS los departamentos (tabla derecha)")
print("• Finanzas aparece aunque no tiene empleados")
print("• Los valores faltantes se llenan con NaN")
print("• Resultado: 5 filas (todos los departamentos)")

print("\n" + "="*60)
```

```python


# 5. COMPARACIÓN VISUAL DE LOS JOINS
print("\n5. COMPARACIÓN VISUAL DE LOS JOINS")

print("INNER JOIN - Solo coincidencias:")
print("Empleados: ✅ Ana, ✅ Carlos, ✅ María, ✅ Pedro, ❌ Laura")
print("Departamentos: ✅ Ventas, ✅ IT, ✅ Marketing, ❌ Finanzas")

print("\nLEFT JOIN - Todos los empleados:")
print("Empleados: ✅ Ana, ✅ Carlos, ✅ María, ✅ Pedro, ✅ Laura")
print("Departamentos: ✅ Ventas, ✅ IT, ✅ Marketing, ❌ Finanzas")

print("\nRIGHT JOIN - Todos los departamentos:")
print("Empleados: ✅ Ana, ✅ Carlos, ✅ María, ✅ Pedro, ❌ Laura")
print("Departamentos: ✅ Ventas, ✅ IT, ✅ Marketing, ✅ Finanzas")

print("\n" + "="*60)
```
