```python
# FILTRADO DE DATOS CON OPERADORES LÓGICOS AND y OR
print("=== FILTRADO CON OPERADORES LÓGICOS AND y OR ===\n")

# 1. LISTA DE EJEMPLO - DATOS DE PERSONAS
print("1. DATOS DE EJEMPLO:")
personas = [
    {"nombre": "Ana", "edad": 25, "ciudad": "Madrid", "salario": 35000, "departamento": "Ventas"},
    {"nombre": "Carlos", "edad": 32, "ciudad": "Barcelona", "salario": 42000, "departamento": "IT"},
    {"nombre": "María", "edad": 28, "ciudad": "Sevilla", "salario": 38000, "departamento": "Ventas"},
    {"nombre": "Pedro", "edad": 35, "ciudad": "Valencia", "salario": 50000, "departamento": "Dirección"},
    {"nombre": "Laura", "edad": 22, "ciudad": "Madrid", "salario": 32000, "departamento": "Marketing"},
    {"nombre": "Javier", "edad": 31, "ciudad": "Barcelona", "salario": 45000, "departamento": "IT"},
    {"nombre": "Sofia", "edad": 27, "ciudad": "Madrid", "salario": 37000, "departamento": "Marketing"},
    {"nombre": "Diego", "edad": 29, "ciudad": "Sevilla", "salario": 41000, "departamento": "IT"}
]

# Mostrar todas las personas
print("Lista completa de personas:")
for i, persona in enumerate(personas, 1):
    print(f"{i}. {persona['nombre']} - {persona['edad']} años - {persona['ciudad']} - ${persona['salario']} - {persona['departamento']}")

print("\n" + "="*70)
```

```python
# 2. FILTRADO CON OPERADOR AND (y)
print("\n2. FILTRADO CON OPERADOR AND (ambas condiciones deben ser True):")

# 2.1 Personas de Madrid Y mayores de 25 años
print("\n2.1 Personas de Madrid Y mayores de 25 años:")
personas_madrid_mayores25 = [p for p in personas if p["ciudad"] == "Madrid" and p["edad"] > 25]
for persona in personas_madrid_mayores25:
    print(f"  - {persona['nombre']} ({persona['edad']} años) - {persona['salario']}")

# 2.2 Personas de IT Y salario mayor a 40000
print("\n2.2 Personas de IT Y salario mayor a $40,000:")
personas_it_alto_salario = [p for p in personas if p["departamento"] == "IT" and p["salario"] > 40000]
for persona in personas_it_alto_salario:
    print(f"  - {persona['nombre']} - ${persona['salario']}")

# 2.3 Personas entre 25 y 30 años Y de Ventas
print("\n2.3 Personas entre 25 y 30 años Y de Ventas:")
personas_25_30_ventas = [p for p in personas if p["edad"] >= 25 and p["edad"] <= 30 and p["departamento"] == "Ventas"]
for persona in personas_25_30_ventas:
    print(f"  - {persona['nombre']} ({persona['edad']} años)")

print("\n" + "="*70)
```

```python
# 3. FILTRADO CON OPERADOR OR (o)
print("\n3. FILTRADO CON OPERADOR OR (al menos una condición debe ser True):")

# 3.1 Personas de Madrid O Barcelona
print("\n3.1 Personas de Madrid O Barcelona:")
personas_madrid_barcelona = [p for p in personas if p["ciudad"] == "Madrid" or p["ciudad"] == "Barcelona"]
for persona in personas_madrid_barcelona:
    print(f"  - {persona['nombre']} - {persona['ciudad']}")

# 3.2 Personas menores de 25 años O salario mayor a 45000
print("\n3.2 Personas menores de 25 años O salario mayor a $45,000:")
personas_jovenes_alto_salario = [p for p in personas if p["edad"] < 25 or p["salario"] > 45000]
for persona in personas_jovenes_alto_salario:
    print(f"  - {persona['nombre']} ({persona['edad']} años) - ${persona['salario']}")

# 3.3 Personas de IT O Marketing
print("\n3.3 Personas de IT O Marketing:")
personas_it_marketing = [p for p in personas if p["departamento"] == "IT" or p["departamento"] == "Marketing"]
for persona in personas_it_marketing:
    print(f"  - {persona['nombre']} - {persona['departamento']}")

print("\n" + "="*70)
```

```python
# 4. COMBINACIÓN AND y OR (con paréntesis para prioridad)
print("\n4. COMBINACIÓN AND y OR (usando paréntesis para claridad):")

# 4.1 Personas de Madrid Y (edad < 25 O salario > 40000)
print("\n4.1 Personas de Madrid Y (edad < 25 O salario > $40,000):")
personas_complejo1 = [p for p in personas if p["ciudad"] == "Madrid" and (p["edad"] < 25 or p["salario"] > 40000)]
for persona in personas_complejo1:
    print(f"  - {persona['nombre']} ({persona['edad']} años) - ${persona['salario']}")

# 4.2 (Personas de IT O Ventas) Y salario > 35000
print("\n4.2 (Personas de IT O Ventas) Y salario > $35,000:")
personas_complejo2 = [p for p in personas if (p["departamento"] == "IT" or p["departamento"] == "Ventas") and p["salario"] > 35000]
for persona in personas_complejo2:
    print(f"  - {persona['nombre']} - {persona['departamento']} - ${persona['salario']}")

# 4.3 Personas de 25-35 años Y (ciudad Madrid O Barcelona) Y no de Dirección
print("\n4.3 Personas de 25-35 años, de Madrid O Barcelona, y no de Dirección:")
personas_complejo3 = [p for p in personas if
                     p["edad"] >= 25 and p["edad"] <= 35 and
                     (p["ciudad"] == "Madrid" or p["ciudad"] == "Barcelona") and
                     p["departamento"] != "Dirección"]
for persona in personas_complejo3:
    print(f"  - {persona['nombre']} ({persona['edad']} años) - {persona['ciudad']} - {persona['departamento']}")

print("\n" + "="*70)
```

```python

# 5. EJEMPLO PRÁCTICO: SISTEMA DE BÚSQUEDA
print("\n5. SISTEMA DE BÚSQUEDA INTERACTIVO:")

def buscar_personas(ciudad=None, edad_min=None, edad_max=None, depto=None, salario_min=None):
    """Función para buscar personas según criterios"""
    resultados = personas

    if ciudad:
        resultados = [p for p in resultados if p["ciudad"] == ciudad]

    if edad_min is not None:
        resultados = [p for p in resultados if p["edad"] >= edad_min]

    if edad_max is not None:
        resultados = [p for p in resultados if p["edad"] <= edad_max]

    if depto:
        resultados = [p for p in resultados if p["departamento"] == depto]

    if salario_min is not None:
        resultados = [p for p in resultados if p["salario"] >= salario_min]

    return resultados

# Ejemplos de búsquedas
print("Búsqueda 1: Personas de Madrid entre 25-35 años:")
resultado1 = buscar_personas(ciudad="Madrid", edad_min=25, edad_max=35)
for p in resultado1:
    print(f"  - {p['nombre']} ({p['edad']} años) - ${p['salario']}")

print("\nBúsqueda 2: Personas de IT con salario > $40,000:")
resultado2 = buscar_personas(depto="IT", salario_min=40000)
for p in resultado2:
    print(f"  - {p['nombre']} - ${p['salario']}")

print("\n" + "="*70)

```

```python
# 6. RESUMEN Y ESTADÍSTICAS
print("\n6. RESUMEN DE FILTRADOS:")

print(f"Total de personas: {len(personas)}")
print(f"Personas usando AND: {len(personas_madrid_mayores25)} ejemplos")
print(f"Personas usando OR: {len(personas_madrid_barcelona)} ejemplos")
print(f"Personas usando combinación AND/OR: {len(personas_complejo1)} ejemplos")

print("\n¡Filtrado completado! ✅")
```
