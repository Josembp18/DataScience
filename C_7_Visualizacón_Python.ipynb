```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Configuración para mejores gráficos
plt.style.use('default')
sns.set_palette("husl")

print("=== VISUALIZACIÓN DE DATOS CON PYTHON - GRÁFICOS BÁSICOS ===\n")

# 1. DATOS DE EJEMPLO (FÁCILES DE ENTENDER)
print("1. CREANDO DATOS DE EJEMPLO...")

# Datos de ventas mensuales
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
ventas_2023 = [120, 150, 180, 160, 200, 190]
ventas_2024 = [130, 160, 170, 190, 210, 220]

# Datos de estudiantes por carrera
carreras = ['Ingeniería', 'Medicina', 'Derecho', 'Arquitectura', 'Administración']
estudiantes = [450, 380, 320, 280, 410]

# Datos de edades para histograma
edades = np.random.normal(22, 3, 200)  # 200 estudiantes con edad promedio 22

# Datos para scatter plot (correlación estudio vs calificación)
horas_estudio = np.random.normal(15, 5, 50)
calificaciones = 60 + horas_estudio * 2 + np.random.normal(0, 5, 50)

print("✅ Datos creados exitosamente!")

print("\n" + "="*60)

# 2. CONFIGURACIÓN DE LA FIGURA PRINCIPAL
print("\n2. CREANDO FIGURA CON TODOS LOS GRÁFICOS...")

fig = plt.figure(figsize=(20, 15))
fig.suptitle('TIPOS DE GRÁFICOS EN PYTHON - CLASE PRÁCTICA',
             fontsize=16, fontweight='bold', y=0.98)

print("\n" + "="*60)

# 3. GRÁFICO DE BARRAS (BAR PLOT)
print("\n3. GRÁFICO DE BARRAS - Comparar categorías")

plt.subplot(3, 3, 1)  # 3 filas, 3 columnas, posición 1
plt.bar(carreras, estudiantes, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FECA57'])
plt.title('ESTUDIANTES POR CARRERA', fontweight='bold')
plt.xlabel('Carreras')
plt.ylabel('Número de Estudiantes')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

# Añadir valores en las barras
for i, v in enumerate(estudiantes):
    plt.text(i, v + 10, str(v), ha='center', va='bottom')

print("✅ Gráfico de barras creado - Perfecto para comparar categorías")

# 4. GRÁFICO DE LÍNEAS (LINE PLOT)
print("\n4. GRÁFICO DE LÍNEAS - Tendencias en el tiempo")

plt.subplot(3, 3, 2)
plt.plot(meses, ventas_2023, marker='o', linewidth=2, label='2023', color='#FF6B6B')
plt.plot(meses, ventas_2024, marker='s', linewidth=2, label='2024', color='#4ECDC4')
plt.title('VENTAS MENSUALES 2023 vs 2024', fontweight='bold')
plt.xlabel('Meses')
plt.ylabel('Ventas (miles)')
plt.legend()
plt.grid(alpha=0.3)

print("✅ Gráfico de líneas creado - Ideal para series de tiempo")

# 5. GRÁFICO CIRCULAR (PIE CHART)
print("\n5. GRÁFICO CIRCULAR - Proporciones porcentuales")

plt.subplot(3, 3, 3)
colores_pie = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FECA57']
plt.pie(estudiantes, labels=carreras, autopct='%1.1f%%', colors=colores_pie, startangle=90)
plt.title('DISTRIBUCIÓN DE ESTUDIANTES', fontweight='bold')

print("✅ Gráfico circular creado - Excelente para mostrar proporciones")

# 6. GRÁFICO DE DISPERSIÓN (SCATTER PLOT)
print("\n6. GRÁFICO DE DISPERSIÓN - Relación entre variables")

plt.subplot(3, 3, 4)
plt.scatter(horas_estudio, calificaciones, alpha=0.7, color='#45B7D1', s=60)
plt.title('HORAS DE ESTUDIO vs CALIFICACIONES', fontweight='bold')
plt.xlabel('Horas de Estudio por Semana')
plt.ylabel('Calificación Final')
plt.grid(alpha=0.3)

# Añadir línea de tendencia
z = np.polyfit(horas_estudio, calificaciones, 1)
p = np.poly1d(z)
plt.plot(horas_estudio, p(horas_estudio), "r--", alpha=0.8)

print("✅ Gráfico de dispersión creado - Muestra correlaciones")

# 7. HISTOGRAMA (HISTOGRAM)
print("\n7. HISTOGRAMA - Distribución de datos")

plt.subplot(3, 3, 5)
plt.hist(edades, bins=15, color='#96CEB4', edgecolor='black', alpha=0.7)
plt.title('DISTRIBUCIÓN DE EDADES DE ESTUDIANTES', fontweight='bold')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.grid(alpha=0.3)

print("✅ Histograma creado - Perfecto para ver distribuciones")

# 8. BOXPLOT (DIAGRAMA DE CAJAS)
print("\n8. BOXPLOT - Resumen estadístico y outliers")

plt.subplot(3, 3, 6)
datos_boxplot = [edades, horas_estudio, calificaciones]
etiquetas_boxplot = ['Edades', 'Horas Estudio', 'Calificaciones']
plt.boxplot(datos_boxplot, labels=etiquetas_boxplot, patch_artist=True)
plt.title('BOXPLOT - COMPARACIÓN DE VARIABLES', fontweight='bold')
plt.ylabel('Valores')
plt.grid(alpha=0.3)

print("✅ Boxplot creado - Muestra mediana, quartiles y outliers")

# 9. GRÁFICO VIOLÍN (VIOLIN PLOT)
print("\n9. GRÁFICO VIOLÍN - Densidad y distribución")

plt.subplot(3, 3, 7)
datos_violin = [edades, horas_estudio]
plt.violinplot(datos_violin, showmeans=True, showmedians=True)
plt.xticks([1, 2], ['Edades', 'Horas Estudio'])
plt.title('VIOLIN PLOT - DENSIDAD DE DISTRIBUCIÓN', fontweight='bold')
plt.ylabel('Valores')
plt.grid(alpha=0.3)

print("✅ Violin plot creado - Combina boxplot con densidad")

# 10. GRÁFICO DE BARRAS HORIZONTAL
print("\n10. GRÁFICO DE BARRAS HORIZONTAL - Alternativa visual")

plt.subplot(3, 3, 8)
plt.barh(carreras, estudiantes, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FECA57'])
plt.title('BARRAS HORIZONTALES', fontweight='bold')
plt.xlabel('Número de Estudiantes')
plt.grid(axis='x', alpha=0.3)

print("✅ Barras horizontales creadas - Útiles para etiquetas largas")

# 11. GRÁFICO DE ÁREA (AREA PLOT)
print("\n11. GRÁFICO DE ÁREA - Tendencias acumulativas")

plt.subplot(3, 3, 9)
meses_num = range(1, 7)
plt.fill_between(meses_num, ventas_2023, alpha=0.6, color='#FF6B6B', label='2023')
plt.fill_between(meses_num, ventas_2024, alpha=0.6, color='#4ECDC4', label='2024')
plt.plot(meses_num, ventas_2023, marker='o', color='#FF6B6B')
plt.plot(meses_num, ventas_2024, marker='s', color='#4ECDC4')
plt.title('GRÁFICO DE ÁREA - VENTAS ACUMULATIVAS', fontweight='bold')
plt.xlabel('Meses')
plt.ylabel('Ventas (miles)')
plt.legend()
plt.grid(alpha=0.3)

print("✅ Gráfico de área creado - Muestra tendencias acumulativas")

print("\n" + "="*60)

# 12. AJUSTAR ESPACIOS Y MOSTRAR GRÁFICOS
print("\n12. AJUSTANDYO Y MOSTRANDO TODOS LOS GRÁFICOS...")

plt.tight_layout()
plt.subplots_adjust(top=0.94)
plt.show()

print("✅ ¡Todos los gráficos mostrados exitosamente!")

print("\n" + "="*60)

# 13. EXPLICACIÓN DE CADA GRÁFICO
print("\n13. EXPLICACIÓN DE CUÁNDO USAR CADA GRÁFICO:")

explicaciones = {
    "Barras": "✅ Comparar categorías discretas (ej: ventas por producto)",
    "Líneas": "✅ Mostrar tendencias en el tiempo (ej: evolución de precios)",
    "Circular": "✅ Mostrar proporciones porcentuales (ej: distribución de mercado)",
    "Dispersión": "✅ Ver relación entre dos variables (ej: altura vs peso)",
    "Histograma": "✅ Analizar distribución de una variable (ej: edades, salarios)",
    "Boxplot": "✅ Resumen estadístico y detectar valores atípicos",
    "Violín": "✅ Combinar distribución y densidad de datos",
    "Barras Horizontal": "✅ Cuando los nombres de categorías son largos",
    "Área": "✅ Mostrar tendencias acumulativas o superpuestas"
}

for grafico, explicacion in explicaciones.items():
    print(f"• {grafico}: {explicacion}")

print("\n" + "="*60)

# 14. CÓDIGO SIMPLIFICADO PARA CADA GRÁFICO
print("\n14. CÓDIGO BÁSICO PARA CADA GRÁFICO:")

codigos = {
    "Barras": "plt.bar(categorias, valores)",
    "Líneas": "plt.plot(x, y)",
    "Circular": "plt.pie(valores, labels=categorias)",
    "Dispersión": "plt.scatter(x, y)",
    "Histograma": "plt.hist(datos, bins=10)",
    "Boxplot": "plt.boxplot(datos)",
    "Violín": "plt.violinplot(datos)"
}

for grafico, codigo in codigos.items():
    print(f"• {grafico}: {codigo}")

print("\n" + "="*60)
print("🎓 ¡CLASE COMPLETADA! Ahora practica con tus propios datos.")
```
