```python
import pandas as pd
import numpy as np

print("=== CREACI√ìN Y EXPLORACI√ìN DE DATAFRAMES CON PANDAS ===\n")

# 1. CREACI√ìN DE DATAFRAMES DESDE DIFERENTES FUENTES
print("1. CREACI√ìN DE DATAFRAMES:")

# 1.1 Desde un diccionario
print("\n1.1 Desde diccionario:")
datos = {
    'Nombre': ['Ana', 'Carlos', 'Mar√≠a', 'Pedro', 'Laura', 'Javier'],
    'Edad': [25, 32, 28, 35, 29, 31],
    'Ciudad': ['Madrid', 'Barcelona', 'Sevilla', 'Valencia', 'Madrid', 'Barcelona'],
    'Salario': [35000, 42000, 38000, 50000, 32000, 45000],
    'Departamento': ['Ventas', 'IT', 'Ventas', 'Direcci√≥n', 'Marketing', 'IT']
}

df = pd.DataFrame(datos)
print("DataFrame creado desde diccionario:")
print(df)


# 1.2 Desde una lista de listas
print("\n1.2 Desde lista de listas:")
columnas = ['Producto', 'Precio', 'Stock', 'Categor√≠a']
datos_productos = [
    ['Laptop', 1200, 15, 'Electr√≥nicos'],
    ['Mouse', 25, 100, 'Electr√≥nicos'],
    ['Silla', 150, 30, 'Oficina'],
    ['Monitor', 300, 20, 'Electr√≥nicos'],
    ['Escritorio', 200, 25, 'Oficina']
]

df_productos = pd.DataFrame(datos_productos, columns=columnas)
print("DataFrame de productos:")
print(df_productos)

print("\n" + "="*70)
```

```python
# 2. EXPLORACI√ìN B√ÅSICA CON HEAD() Y TAIL()
print("\n2. EXPLORACI√ìN CON HEAD() Y TAIL():")

print("Primeras 3 filas con head(3):")
print(df.head(3))

print("\n√öltimas 2 filas con tail(2):")
print(df.tail(2))

print("\nPor defecto head() muestra 5 filas:")
print(df.head())

print("\nPor defecto tail() muestra 5 filas:")
print(df.tail())

print("\n" + "="*70)
```

```python
# 3. INFORMACI√ìN ESTRUCTURAL CON SHAPE
print("\n3. INFORMACI√ìN ESTRUCTURAL CON SHAPE:")

print("DataFrame principal:")
print(f"df.shape: {df.shape}")
print(f"‚Üí {df.shape[0]} filas y {df.shape[1]} columnas")

print("\nDataFrame de productos:")
print(f"df_productos.shape: {df_productos.shape}")
print(f"‚Üí {df_productos.shape[0]} filas y {df_productos.shape[1]} columnas")

print("\n" + "="*70)
```

```python
# 4. INFORMACI√ìN DETALLADA CON INFO()
print("\n4. INFORMACI√ìN DETALLADA CON INFO():")

print("Informaci√≥n del DataFrame principal:")
print("df.info():")
df.info()

print("\nInformaci√≥n del DataFrame de productos:")
print("df_productos.info():")
df_productos.info()

print("\n" + "="*70)
```

```python
# 5. M√âTODOS ADICIONALES DE EXPLORACI√ìN
print("\n5. M√âTODOS ADICIONALES DE EXPLORACI√ìN:")

print("5.1 Columnas disponibles:")
print(f"df.columns: {list(df.columns)}")

print("\n5.2 Tipos de datos por columna:")
print("df.dtypes:")
print(df.dtypes)

print("\n5.3 Estad√≠sticas descriptivas:")
print("df.describe():")
print(df.describe())

print("\n5.4 Valores √∫nicos en columna 'Ciudad':")
print(f"df['Ciudad'].unique(): {df['Ciudad'].unique()}")

print("\n5.5 Conteo de valores en columna 'Departamento':")
print("df['Departamento'].value_counts():")
print(df['Departamento'].value_counts())

print("\n" + "="*70)
```

```python
# 6. DATAFRAME CON DATOS M√ÅS COMPLEJOS
print("\n6. DATAFRAME CON DATOS M√ÅS COMPLEJOS:")

# Crear DataFrame con diferentes tipos de datos y valores nulos
datos_complejos = {
    'ID': [1, 2, 3, 4, 5, 6],
    'Nombre': ['Ana', 'Carlos', None, 'Pedro', 'Laura', 'Javier'],
    'Edad': [25, 32, 28, None, 29, 31],
    'Email': ['ana@email.com', 'carlos@email.com', 'maria@email.com',
              'pedro@email.com', None, 'javier@email.com'],
    'Fecha_Contratacion': pd.date_range('2023-01-01', periods=6, freq='M'),
    'Activo': [True, True, False, True, True, False],
    'Puntuacion': [85.5, 92.3, 78.9, None, 88.1, 95.7]
}

df_complejo = pd.DataFrame(datos_complejos)
print("DataFrame con datos complejos:")
print(df_complejo)

print("\nInformaci√≥n con info():")
df_complejo.info()

print("\nValores nulos por columna:")
print(df_complejo.isnull().sum())

print("\n" + "="*70)
```

```python
# 7. EJEMPLO PR√ÅCTICO COMPLETO
print("\n7. EJEMPLO PR√ÅCTICO: AN√ÅLISIS DE EMPLEADOS")

print("=== RESUMEN DEL AN√ÅLISIS ===")
print(f"Total de empleados: {len(df)}")
print(f"Ciudades representadas: {df['Ciudad'].nunique()}")
print(f"Departamentos: {', '.join(df['Departamento'].unique())}")

print(f"\nSalario promedio: ${df['Salario'].mean():.2f}")
print(f"Salario m√°ximo: ${df['Salario'].max()}")
print(f"Salario m√≠nimo: ${df['Salario'].min()}")

print(f"\nDistribuci√≥n por ciudad:")
print(df['Ciudad'].value_counts())

print(f"\nDistribuci√≥n por departamento:")
print(df['Departamento'].value_counts())

print("\n¬°Exploraci√≥n completada! üêº")
```
